```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 3]
```

```python
def displayImages(img1,img2):
    fig= plt.figure(figsize=(10,20))
    fig.add_subplot(1, 2, 1)
    plt.imshow(img1, cmap='gray')
    plt.axis('off')
    plt.title("First")
    
    fig.add_subplot(1, 2, 2)
    plt.imshow(img2, cmap='gray')
    plt.axis('off')
    plt.title("Second")
    plt.show()
```

```python
image=cv2.imread('downloads/threshold.jpg',0)
array=np.array(image)
threshold=70
new_array=np.where(array<threshold,0,array)
new_image=cv2.convertScaleAbs(new_array)
displayImages(image,new_image)
```
