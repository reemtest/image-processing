```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 3]
```

```python
def displayImage(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
img_1= cv2.imread('./downloads/concat1.jpg',0)
cv2.imshow("concat1", img_1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_2= cv2.imread('./downloads/concat2.jpg',0)
cv2.imshow("concat2", img_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_3= cv2.imread('./downloads/concat3.jpg',0)
cv2.imshow("concat3", img_3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_2 = cv2.resize(img_2, (img_1.shape[1], img_1.shape[0]))
img_3 = cv2.resize(img_3, (img_1.shape[1], img_1.shape[0]))
```

```python
numpydata_1 = np.asarray(img_1)
numpydata_2 = np.asarray(img_2)
numpydata_3 = np.asarray(img_3)
```

```python
new_array=np.concatenate((numpydata_1, numpydata_2,numpydata_3), axis=1)
new_image=cv2.convertScaleAbs(new_array)
displayImage(new_image)
```
